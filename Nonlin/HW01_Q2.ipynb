```python
import csv
import numpy as np
from matplotlib import pyplot as plt
import hsic
from sklearn.svm import SVR
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.np.array(df.x)read_csv('dag.csv')
X = 
Y = np.array(df.y)
Z = np.array(df.z)
W = np.array(df.w)

X = np.reshape(X,newshape=(len(X), 1))
Y = np.reshape(Y,newshape=(len(Y), 1))
Z = np.reshape(Z,newshape=(len(Z), 1))
W = np.reshape(W,newshape=(len(W), 1))
```

```python
# W -----> X
clf = SVR(kernel='rbf')
clf.fit(W,X.ravel())


# sorting idx for plot
idx   = np.argsort(W.ravel())
X_bar =  clf.predict(W[idx])


plt.figure()
axis = plt.axes()
axis.scatter(W, X,color = 'b')
axis.plot(W[idx], X_bar, color='r')
axis.set_xlabel('W')
axis.set_ylabel('X')
axis.set_title('W---->X')

X_bar = clf.predict(W).reshape(len(X_bar),1)
residual = X_bar - X

plt.figure()
axis = plt.axes()
axis.scatter(W[idx],residual,color = 'b')
axis.set_xlabel('W')
axis.set_ylabel('Residual_X')
axis.set_title('Residual')

Output = hsic.hsic_gam(residual, W,  0.02)
print(Output)
```
